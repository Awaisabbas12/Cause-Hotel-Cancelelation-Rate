# Removing Outlier

```python
df.describe()
```

```python
df.info()
```

```python
def remove_outlier(data):
    Q1 = data.iloc[(len(data)*25//100)-1]
    Q3 = data.iloc[(len(data)*75//100)-1]
    IQR = Q3- Q1
    lf = Q1 - 1.5*IQR
    uf = Q3 + 1.5*IQR
    return (data[(data>=lf) & (data<=uf)])
```

```python
df['adults'] = remove_outlier(df['adults'])
df['children'] = remove_outlier(df['children'])
df['babies']=remove_outlier(df['babies'])
df['is_repeated_guest']=remove_outlier(df['is_repeated_guest'])
df['previous_cancellations']=remove_outlier(df['previous_cancellations'])
df['days_in_waiting_list']=remove_outlier(df['days_in_waiting_list'])
df['adr']=remove_outlier(df['adr'])
```

# Data analysis and data visualization

```python
canceled_perC = df['is_canceled'].value_counts(normalize = True)# normalize = true give us in perccent
print(canceled_perC)
```

```python
canceled_perC = df['is_canceled'].value_counts(normalize = True)
print(canceled_perC)
plt.figure(figsize = (4,4))
plt.title('Reservation_status count')
plt.bar(['not_canceled','canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.8)
plt.show()
```

```python
plt.figure(figsize = (5,4))
ax1 = sbn.countplot(x='hotel',hue ='is_canceled',data=df,palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('number of reservations in different hotel')
plt.xlabel('hotels')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()#adr means average daily rate
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel
```

```python
plt.figure(figsize = (20,8))
plt.title("Average daily rate in resort and city hotels",size = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'resort_hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'city_hotel')
plt.legend()
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (14,6))
ax1= sbn.countplot(x='month',hue = 'is_canceled',data = df)
legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('number of reservations per month')
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.bar('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('Average Daily rate per month')
plt.ylabel('Average Daily Rate')
plt.show()
```
